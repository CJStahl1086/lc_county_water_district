```python
import pandas as pd
import numpy as np
```

```python
#importing surface water data
stream = pd.read_excel('Stream.xlsx')
stream.head(3)
```

```python
# splitting data frame to merge
stream1 = stream[['site_id','collection_date','collection_time','matrix','measured_by','agency', 'notes']]
stream1 = stream1.drop_duplicates()
stream1
```

```python
# preparing to merge the transposed column of characteristic name
stream_merge = stream[['site_id','collection_date', 'characteristic_name', 'result_value']]
stream_merge = stream_merge.pivot(index=['site_id','collection_date'], columns='characteristic_name', values='result_value')
stream_merge = stream_merge.reset_index()
stream_merge
```

```python
#merging stream_merge dataframe to stream data
final = pd.merge(stream_merge, stream1, on=['site_id', 'collection_date'], how='left')
final = final.rename(columns={'Water Temperature' : 'water_temp_c', 'pH' : 'ph', 'Dissolved Oxygen Concentration' : 'dissolved_oxygen_mgl',
                     'Dissolved Oxygen Saturation' : 'dissolved_oxygen_percent', 'Specific Conductivity' : 'spc_uscm', 'Total Dissolved Solids' : 'tsd_mgl',
                     'Salinity': 'salinity_ppt', 'Conductivity':'conductivity'})
final
```

```python
#importing groundwater for final merge of ground water and surface water data sets
gw = pd.read_excel('Groundwater.xlsx')

FINAL = pd.concat([final, gw], axis=0)
FINAL.to_csv('field_params.csv')
FINAL.head(5)
```
