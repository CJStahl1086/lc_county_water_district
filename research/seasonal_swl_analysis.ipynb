```python
import pandas as pd
```

```python
maddie = pd.read_csv('swl_analysis.csv')
maddie = maddie.drop(labels=['Unnamed: 0'], axis=1)
summer = maddie[maddie['season'] == 'summer'].copy()
winter = maddie[maddie['season'] == 'winter'].copy()
fall = maddie[maddie['season'] == 'fall'].copy()
spring = maddie[maddie['season'] == 'spring'].copy()
```

```python
#grouping by each time period to transpose into into columns
summer1 = summer.groupby(['gwicid','time_period','swl_ground','change'])['latitude'].count().reset_index()
summer1 = summer1.pivot(index='gwicid',columns=['time_period'], values='swl_ground').reset_index()
```

```python
#creating another group by object to merge into final format for ArcGIS with lat-long data
summer = summer[['gwicid', 'latitude','longitude','count','change']].copy()
summer = summer.groupby(['gwicid','latitude','longitude'])['count'].mean().reset_index()
summer1 = summer1.merge(summer, on='gwicid', how='left')
summer1 = summer1.drop(labels='count', axis=1)
summer1.to_csv('summer.csv')
summer1
```

```python
#creating another group by object to merge into final format for ArcGIS with lat-long data
winter = winter[['gwicid', 'latitude','longitude','count']].copy()
winter = winter.groupby(['gwicid','latitude','longitude'])['count'].mean().reset_index()
winter1 = winter1.merge(summer, on='gwicid', how='left')
winter1 = winter1.drop(labels='count', axis=1)
winter1.to_csv('winter.csv')
winter1
```
