```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
indmerge = pd.read_excel('Ind_Residents_Merged.xlsx')
indmerge.head()
```

```python
#checking for duplicates
indmerge.duplicated(subset=['address','collection_date','characteristic_name','result_value','result']).sum()
```

```python
#capitalizing all values
indmerge = indmerge.apply(lambda x: x.astype(str).str.upper())

#converting to datetime mm/dd/yyyy format
indmerge['collection_date'] = pd.to_datetime(indmerge['collection_date']).dt.strftime('%m/%d/%Y')
indmerge['collection_date'] = pd.to_datetime(indmerge['collection_date'])

#convert dtypes to float
indmerge[['result_value','lat','long','reporting_limit']] = indmerge[['result_value','lat','long','reporting_limit']].apply(pd.to_numeric, errors='coerce')

#updating program values from notes
indmerge['program'] = np.where(indmerge['notes'].str.contains('WELL EDUCATED'), 'WELL EDUCATED', indmerge['program'])
indmerge.head()
```

```python
indmerge.to_excel('ind_resident_final.xlsx')
```
