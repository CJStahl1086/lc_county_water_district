```python
import datetime as dt
import pandas as pd 
import numpy as np
```

```python
# Cleaning of orignal data

master = pd.read_excel('All WQ Data from Access.xlsx', 'Sheet1')
master.columns = [x.lower() for x in master.columns]
master.columns = master.columns.str.replace(' ', '_', regex=True)
master = master.rename(columns={'projectname': 'project_name','collectiontime': 'collection_time', 'collectionyear':
                               'collection_year'})
master.to_csv('MasterDF.csv')
```

```python
# Added arsenic data manually (852 rows) 

master = pd.read_csv('MasterDF.csv')
arsenic = pd.read_csv('Arsenic 2022 Data.csv')
arsenic.columns = [x.lower() for x in arsenic.columns]
arsenic = arsenic.rename(columns={'#activity_id':'activity_id', 'collection date': 'collection_date',
                                 'detection_limit_units': 'detection_limit_unit', 'result_comment':'notes'})
final = pd.concat([master, arsenic], axis=0)
final.to_csv('MasterDF.csv')
```

```python
# Creating function to clean and combine data to master.csv

def clean(df):
    #bring in master data to append new cleaned to
    master = pd.read_csv('MasterDF.csv')
    master = master.drop(columns='Unnamed: 0')
    
    #creating snake case
    df.columns = [x.lower() for x in df.columns]
    df.columns = df.columns.str.replace(' ', '_', regex=True)
    
    #renaming columns
    df = df.rename(columns={'result_comment':'notes','#activity_id': 'activity_id'})
    
    #extracting project name
    df['notes'] = df['notes'].astype(str)
    df['project_name'] = df['notes'].str.split('Project Name:').str[1]
    
    #extracting/formatting dates
    df['Time stamp'] = pd.to_datetime(df['collection_date'])
    df['collection_date'] = df['collection_date'].str.split(' ').str[0]
    df['hour'] = df['Time stamp'].dt.hour
    df['minute'] = df['Time stamp'].dt.minute
    df['collection_time'] = df['hour'].astype(str) + ':' + df['minute'].astype(str)
    df['collection_year'] = df['Time stamp'].dt.year
    
    #deleting columns not found in master dataframe
    delcols = ['data_logger_line_id', 'biological_intent', 'biological_taxonomic_name', 'biological_unidentified_species_name',
               'biological_tissue_anatomy_name','characteristic_id','lab_sample_prep_method_id','laboratory_comment_code',
               'result_sampling_point', 'result_depth_height_measure', 'result_depth_height_measure_unit',
               'statistic_type', 'weight_basis', 'particle_size_basis', 'precision_value', 'result_file_name',
               'Time stamp','hour', 'minute']
    df = df.drop(columns=delcols)
    
    #add to current Master dataset and overwrite currentmasterfile
    final = pd.concat([master, df], axis=0)
    final = final.reset_index(drop=True)
    final.to_csv('MasterDF.csv')
    newfinal = pd.read_csv('MasterDF.csv')
    return(newfinal)
```

```python
#importing all Energy lab data set and updating the master list sequentially with above function ('clean')

cole = pd.read_csv('203 Coleman.csv')
clean(cole)  #20 rows verified master updated

wins= pd.read_csv('1247 Winscott Ln.csv')
clean(wins)   #4 rows verified master updated

NH = pd.read_csv('04132022_NH.csv')
clean(NH)   #72 rows verified master updated

NH1 = pd.read_csv('04162022_NH.csv')
clean(NH1) #72 rows verified master updated

NH2 = pd.read_csv('05112021_NH.csv')
clean(NH2)  #19 rows verified master updated

NH3 = pd.read_csv('06072022_NH.csv')
clean(NH3) #231 rows verified master updated

NH4 = pd.read_csv('07142022_NH.csv')
clean(NH4) #36 rows verified master updated

betty = pd.read_csv('Betty Sharp.csv')
clean(betty) #17 rows verified master updated

forest = pd.read_csv('Forestvale EDD.csv')
clean(forest) #31 rows verified master updated

spokane = pd.read_csv('Spokane_Creek.csv')
clean(spokane) #92 rows verified master updated

tmc = pd.read_csv('TMC.csv')
clean(tmc)       #200 rows verified master updated

```
