```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import seaborn as sns
```

```python
def clean(df):
    df['Date, GMT-06:00'] = pd.to_datetime(sevenmile['Date, GMT-06:00'])
    df.rename(columns={ df.columns[2]: 'Temp' }, inplace = True)
    a = df.groupby(pd.Grouper(key='Date, GMT-06:00', axis=0, 
                      freq='D'))[['discharge_values_cfs']].mean().reset_index()
    b = df.groupby(pd.Grouper(key='Date, GMT-06:00', axis=0, 
                      freq='D')).agg({'Temp': ['min','max']}).reset_index()
    b.columns = b.columns.droplevel()
    b = b.rename(columns={'':'Date, GMT-06:00','max':'temps_max', 'min':'temps_min'})
    c = a.merge(b, on=['Date, GMT-06:00'])
    sns.lineplot(data=c, x ='Date, GMT-06:00', y = 'discharge_values_cfs')
    return(c)
```

```python
sevenmile = pd.read_csv('7M-1_SevenMile_Creek_at_Birdseye_Water_Level.csv')
a = clean(sevenmile)
a.to_csv('7M-1_SevenMile_Creek_at_Birdseye_Water_Level_clean.csv')
```
